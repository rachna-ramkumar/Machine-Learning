# Decision Tree

```python
import numpy as np
import matplotlib.pyplot as plt
import graphviz
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
```

```python
iris = datasets.load_iris()
X = iris.data 
y = iris.target 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

# Gini and Entropy with Depth 1

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_text
iris = load_iris()
decision_tree = DecisionTreeClassifier(random_state=0, max_depth=1)
decision_tree = decision_tree.fit(iris.data, iris.target)
r = export_text(decision_tree, feature_names=iris['feature_names'])
print(r)
```

```python
gini = DecisionTreeClassifier(criterion='gini',max_depth=1,random_state=1)
gini.fit(X_train, y_train)

print('Gini Depth 1 Mean Accuracy:', gini.score(X_test, y_test))

export_graphviz(gini, out_file='gini5.dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('gini5.dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)



```

```python
entropy = DecisionTreeClassifier(criterion='entropy',max_depth=1,random_state=1)
entropy.fit(X_train, y_train)

print('Entropy Depth 1 Mean Accuracy:', entropy.score(X_test, y_test))

export_graphviz(entropy, out_file='entropy5dot', feature_names=['sepal length','sepal width','petal length','petal width'])
with open('entropy5dot') as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

# Discussion
 
Gini impurity measures how heterogeneous or combined some value is over a set When evaluating Gini impurity and entropy, it’s essential to take into account that the greater the rating the higher the impurity, and therefore the greater the statistics acquired will be when a split occurs.

For this Assignement, I Ran the Decision Tree model for trees of depth 1, 2, 3, 4, 5, and 6 and for the Gini Impurity and Entropy impurity measures for each tree depth and got results which are mentioned above.

We can conclude that gini and entropy impurity measures are negligible in difference and are pretty much identical, for this dataset at least since the accuracy from both impurity measures are the same at the same tree depth.
