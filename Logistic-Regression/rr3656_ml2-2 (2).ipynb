Non Linearly Seperable Data (2 Features and 2 Classes)

```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn import svm
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

iris = datasets.load_iris()
df = pd.DataFrame(iris.data)
df['target'] = iris.target
df = df[(df['target'] == 2)|(df['target'] == 0)]
X = df[[0, 1, 2 , 3]].values
Y = df['target']

#THe part above is common for all 
#The part below is logistic regression part
# Penalty can be a few (l1/l2) is what is mentioned in your assignment 
# C is how much you want to regularize - as its an inverse, smaller the value the more the regularization 
# If you remove the C value and train, due to overfitting you will always get accuray of 1 (as with the other 3 below)
# https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
logisticRegr = LogisticRegression(penalty = 'l2', C = 0.00000000001)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.24, random_state=0)
logisticRegr.fit(x_train, y_train)
predictions = logisticRegr.predict(x_test)
print(predictions)
score = logisticRegr.score(x_test, y_test)
print(score)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA

iris = datasets.load_iris()
X = iris.data[:, [0, 1]] 
y = iris.target

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(2, figsize=(8, 6))
plt.clf()


plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(X_reduced[:, 0], X_reduced[:, 1], X_reduced[:, 2], c=y,
           cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
